## Pytorch

```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
from torchsummary import summary


class AlexNet(nn.Module):

    def __init__(self, n_classes):
        super(AlexNet, self).__init__()
        
        ############## Add Layer ##############
        self.conv1 = nn.Sequential(nn.Conv2d(3, 96, 11, stride=4, padding=2),
                                   nn.ReLU(),
                                   nn.MaxPool2d(2,2))    
        self.conv2 = nn.Sequential(nn.Conv2d(96, 256, 5, padding='same'),
                                   nn.ReLU(),
                                   nn.MaxPool2d(2,2))    
        self.conv3 = nn.Sequential(nn.Conv2d(256, 384, 3,  padding='same'),
                                   nn.ReLU())    
        self.conv4 = nn.Sequential(nn.Conv2d(384, 384, 3,  padding='same'),
                                   nn.ReLU())    
        self.conv5 = nn.Sequential(nn.Conv2d(384, 256, 3,  padding='same'),
                                   nn.ReLU(),
                                   nn.MaxPool2d(2,2))    
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(9216, 4096),
            nn.ReLU(),
            nn.Linear(4096,4096),
            nn.ReLU(),
            nn.Linear(4096, n_classes))
        #######################################

    def forward(self, x):
        
        ############## Add Layer ##############
        x = self.conv1(x)    
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.conv5(x)
        print(x.shape)
        x = self.fc(x)

        #######################################
        
        return x

model = AlexNet(1000)
# pytorch summary
summary(model, input_size=(3, 224, 224), device='cpu') # summary code 추가
```
